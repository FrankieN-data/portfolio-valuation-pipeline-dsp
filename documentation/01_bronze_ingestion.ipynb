### Raw to Bronze Transformations

**01_bronze_csv_to_dlt.py**

1. Maintain 1:1 Source Fidelity
2. Enforce Append-Only Storage (Historical Ledger)
3. Append Lightweight Operational Metadata (Audit)
5. Schema Evolution Isolation

- Loading csv files into Delta Lake Table using hybrid approach: enforcement of basic data types (integer, double, string). Goal is to limit the number of rows rejected while offering best performance for streaming and high-volume.
- Each inserted with a timestamp, source name and source system name for audit and SCD purpose. 
- Systematic logic implemented using DRY principle

**Tactical decisions**

Loading strategy is a batch loading - more adequate to develop and test home project. To transition to streaming ingestion. Modifications include:

- Replacing `spark.read.format("csv")` with `spark.readStream.format("cloudFiles").option("cloudFiles.format", "csv")`

**Technical Decisions**

- **Storage**: Parquet format was a good choice for this use case - column storage less space, quick, one-time loading for home project, read-only mainly scenario once pipeline run once. Using DLT to showcase skills. 
- **Transformation**: PySpark over SQL, avoid code deduplication, implemented a mechanism that can be extended to more table more easily (set the file path, define the schema, add the parameters tuple, a one-line command to create the table).
- **Governance**: Code versioning in GitHub, CI/CD with tests: configuration, DAB for deployment 

Built a schema-on-read bronze ingestion pipeline
